```python
from selenium import webdriver
```

```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
import datetime
import time

from requests_html import HTML
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import WebDriverException

from webdriver_manager.chrome import ChromeDriverManager
from requests_html import HTML
```

```python
browser = web_driver()
browser.get('https://www.worldweatheronline.com/da-nang-weather-history/vn.aspx')

f = open('weatherDaNang.csv', 'w', encoding='utf-8')
f.write('Date\tWeather\tTime\tTemperature\tForecast\tRain\t%Rain\tCloud\tPressure\tWind\tGust\n')

date = datetime.datetime(2009, 1, 1)
end_date = datetime.datetime(2024, 11, 30)
date_col = ''

record = ['Weather', 'Time', 'Temperature', 'Forecast', 'Rain', '% Rain', '% Cloud', 'Pressure', 'Wind', 'Gust']
n_cols = len(record)

while date < end_date:
    date_str = date.strftime('%Y-%m-%d')
    # date_str = date.strftime('%d/%m/%Y')

    input_date = browser.find_element(By.ID,'ctl00_MainContentHolder_txtPastDate')
    browser.execute_script("arguments[0].value = arguments[1];", input_date, date_str)
    submit_date = browser.find_element(By.ID,'ctl00_MainContentHolder_butShowPastWeather')

    try:
        submit_date.click()
    except WebDriverException:
        print(f'Error: Date {date_str}')
        continue

    time.sleep(1)

    yesterday = date_col
    date_col = browser.find_element(By.CLASS_NAME, 'days-collapse-date').text
    if date_col == yesterday:
        continue

    date += datetime.timedelta(days=1)

    rows =  browser.find_elements(By.CLASS_NAME, 'days-details-row')[1:9]
    weathers = browser.find_elements(By.CLASS_NAME, 'days-table-forecast-img')[0:8]

    j = 0
    for row in rows:
      i = 0
      data = row.text.splitlines()

      while i < n_cols :
        if i == 5:
            temp = data[i].split(' ')
            data[i - 1] = temp[0]
            data[i] = temp[1]
            data[i + 1] = temp[2]
        elif i == 6:
            data[i] = data[i] + ' mb'
        elif i == 4 :
            data[i - 1] = data[i - 1] + ' mm'
        i += 1

      if i == n_cols:
          weather = weathers[j].get_attribute('title')
          j += 1
          f.write(f'{date_col}\t{weather}\t{data[0]}\t{data[1]}\t{data[2]}\t{data[3]}\t{data[4]}\t{data[5]}\t{data[6]}\t{data[7]}\t{data[8]}\n')
    print(date_col)

f.close()
```
